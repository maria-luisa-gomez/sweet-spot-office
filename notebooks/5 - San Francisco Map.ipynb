#### MAP v.2

```python
map_2 = folium.Map(location=[inicial_lat, inicial_long], zoom_start=13)
```

```python
icono_office = Icon(color = "orange",
             opacity=1,
             prefix = "fa",
             icon = "fa-heart",
             icon_color = "black",
)

loc = {"location" : [inicial_lat, inicial_long], "tooltip":"GAMING COMPANY"}

#guardamos el marker en una variable
marker_office = folium.Marker(**loc, icon = icono_office)
marker_office.add_to(map_2)
```

'Preschool', 'Going Out', 'Vegan Restaurant', 'Transportation', 'Pet Services'

```python
for i, row in df_san_francisco_map.iterrows():
    loc = {
        "location": [row["lat"], row["long"]], 
        "tooltip": row["name"]
    }
    
    if row["category_simplify"] == "Preschool":
        icono = Icon( color="blue",
               prefix="fa",
               icon="fa-child",
               icon_color="black")
    
    elif row["category_simplify"] == "Going Out":
        icono = Icon( color="cadetblue",
               prefix="fa",
               icon="fa-glass",
               icon_color="black")
        
    elif row["category_simplify"] == "Vegan Restaurant":
        icono = Icon( color="green",
               prefix="fa",
               icon="fa-lemon-o",
               icon_color="black")
        
    elif row["category_simplify"] == "Transportation":
        icono = Icon( color="beige",
               prefix="fa",
               icon="fa-taxi",
               icon_color="black")
    else: 
        icono = Icon( color="red",
               prefix="fa",
               icon="fa-scissors",
               icon_color="black")
        
    mar = Marker(**loc, icon=icono)
    mar.add_to(map_2)
map_2
```
